### Correlacion: Deforestacion vs. Especies en peligro
Se analiza si existe relacion entre la perdida de cobertura forestal y el numero de especies en peligro en Honduras a lo largo del tiempo.

```python
cols_loss = [col for col in df_deforest.columns if 'tc_loss_ha_' in col]
df_loss_hnd = df_deforest[df_deforest['threshold'] == 0][cols_loss].sum()
df_loss_hnd.index = df_loss_hnd.index.str.replace('tc_loss_ha_', '')
df_loss_hnd = df_loss_hnd.sort_index()
years = df_loss_hnd.index.astype(int)
total_loss = df_loss_hnd.values

especies_hnd = df_especies[df_especies['Name'] == 'Honduras'].iloc[0]

total_especies = int(especies_hnd['Total'])

print(f"ANALISIS DE CORRELACION: DEFORESTACION vs. ESPECIES EN PELIGRO")
print(f"Total de especies en peligro en Honduras (UICN): {total_especies}")
print(f"Periodo de analisis de deforestacion: {min(years)}-{max(years)}")
print(f"Perdida total de cobertura forestal en el periodo: {sum(total_loss):,.0f} hectareas")

plt.figure(figsize=(15,10))

plt.subplot(3,1,1)
plt.bar(years, total_loss, alpha=0.7, color='red')
plt.title('Perdida anual de cobertura forestal en Honduras', fontsize=14, fontweight='bold')
plt.xlabel('Ano')
plt.ylabel('Hectareas perdidas')
plt.grid(True, alpha=0.3)

plt.subplot(3,1,2)
grupos = ['Mammals','Birds','Reptiles*','Amphibians','Fishes*','Molluscs*','Other Inverts*','Plants*']
valores_grupos = []
for grupo in grupos:
    if grupo in especies_hnd.index:
        val = especies_hnd[grupo]
        if isinstance(val, str) and ',' in val:
            val = int(val.replace(',',''))
        else:
            val = int(val) if pd.notna(val) else 0
        valores_grupos.append(val)
    else:
        valores_grupos.append(0)

plt.bar(grupos, valores_grupos, alpha=0.7, color='orange')
plt.title('Especies en peligro por grupo taxonomico en Honduras', fontsize=14, fontweight='bold')
plt.ylabel('Numero de especies')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)

plt.subplot(3,1,3)
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(years, total_loss)
trend_line = slope * years + intercept

plt.plot(years, total_loss, 'ro-', linewidth=2, markersize=6, label='Deforestacion anual')
plt.plot(years, trend_line, 'r--', alpha=0.7, label=f'Tendencia (pendiente: {slope:.0f} ha/ano)')
plt.axhline(y=np.mean(total_loss), color='blue', linestyle=':', alpha=0.7, 
           label=f'Promedio: {np.mean(total_loss):,.0f} ha/ano')
plt.title('Tendencia de deforestacion en Honduras', fontsize=14, fontweight='bold')
plt.xlabel('Ano')
plt.ylabel('Hectareas perdidas')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\nANALISIS ESTADISTICO DE DEFORESTACION")
print(f"Tendencia de deforestacion: {slope:.2f} ha/ano")
print(f"Coeficiente de correlacion temporal: {r_value:.3f}")
print(f"Significancia estadistica: p = {p_value:.4f}")
print(f"Deforestacion promedio anual: {np.mean(total_loss):,.0f} ha")
print(f"Ano con mayor deforestacion: {years[np.argmax(total_loss)]} ({max(total_loss):,.0f} ha)")
print(f"Ano con menor deforestacion: {years[np.argmin(total_loss)]} ({min(total_loss):,.0f} ha)")

print(f"\nANALISIS DE RELACION CONCEPTUAL")
deforest_rate = slope
if deforest_rate > 0:
    print(f"La deforestacion muestra una tendencia CRECIENTE de {abs(deforest_rate):.0f} ha/ano")
    print("   Esto podria contribuir al aumento de especies en peligro por:")
    print("   • Perdida de habitat")
    print("   • Fragmentacion de ecosistemas")
    print("   • Reduccion de corredores biologicos")
else:
    print(f"La deforestacion muestra una tendencia DECRECIENTE de {abs(deforest_rate):.0f} ha/ano")
    print("   Esto es positivo para la conservacion de especies")

print(f"\nGRUPOS TAXONOMICOS MAS AFECTADOS")
grupos_ordenados = sorted(zip(grupos, valores_grupos), key=lambda x: x[1], reverse=True)
for i, (grupo, cantidad) in enumerate(grupos_ordenados[:5]):
    print(f"{i+1}. {grupo}: {cantidad} especies en peligro")

print(f"\nCONCLUSION: Aunque no podemos calcular una correlacion temporal directa")
print(f"   (los datos de especies son un snapshot), el analisis muestra que Honduras")
print(f"   tiene {total_especies} especies en peligro y una tendencia de deforestacion")
print(f"   {'creciente' if slope > 0 else 'decreciente'}, lo que {'podria agravar' if slope > 0 else 'podria mejorar'}")
print(f"   la situacion de conservacion de biodiversidad.")
```

### Correlacion: Emisiones de CO₂ vs. Temperatura
Se calcula la correlacion y se visualizan juntas las series de emisiones de CO₂ y temperatura promedio en Honduras.

```python
co2_annual = df_co2[df_co2['country']=='Honduras'].set_index('year')['co2']
temp_annual = df_temp_anual.set_index('Year')['Estimated_AverageTemperature_C']

df_corr = pd.DataFrame({'co2': co2_annual, 'temp': temp_annual})
df_corr = df_corr.dropna()

print(f"ANALISIS DE CORRELACION: EMISIONES CO₂ vs. TEMPERATURA")
print(f"Periodo de analisis: {df_corr.index.min()}-{df_corr.index.max()}")
print(f"Numero de anos con datos completos: {len(df_corr)}")
print(f"Rango de emisiones CO₂: {df_corr['co2'].min():.2f} - {df_corr['co2'].max():.2f} Mt")
print(f"Rango de temperaturas: {df_corr['temp'].min():.2f} - {df_corr['temp'].max():.2f}°C")

plt.figure(figsize=(18,12))

plt.subplot(2,3,1)
ax1 = plt.gca()
ax2 = ax1.twinx()

line1 = ax1.plot(df_corr.index, df_corr['co2'], 'b-', linewidth=3, marker='o', 
                markersize=6, label='Emisiones CO₂ (Mt)')
ax1.set_xlabel('Ano')
ax1.set_ylabel('CO₂ (Mt)', color='b', fontweight='bold')
ax1.tick_params(axis='y', labelcolor='b')

line2 = ax2.plot(df_corr.index, df_corr['temp'], 'r-', linewidth=3, marker='s', 
                markersize=6, label='Temperatura (°C)')
ax2.set_ylabel('Temperatura (°C)', color='r', fontweight='bold')
ax2.tick_params(axis='y', labelcolor='r')

plt.title('Series temporales CO₂ y Temperatura', fontsize=14, fontweight='bold')
ax1.grid(True, alpha=0.3)

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

plt.subplot(2,3,2)
from scipy.stats import pearsonr
from sklearn.linear_model import LinearRegression
import numpy as np

corr, pval = pearsonr(df_corr['co2'], df_corr['temp'])

X = df_corr['co2'].values.reshape(-1,1)
y = df_corr['temp'].values
reg = LinearRegression().fit(X, y)
y_pred = reg.predict(X)

plt.scatter(df_corr['co2'], df_corr['temp'], s=80, alpha=0.7, c='purple')
plt.plot(df_corr['co2'], y_pred, 'r-', linewidth=2, 
         label=f'Regresion (R² = {reg.score(X,y):.3f})')
plt.xlabel('Emisiones CO₂ (Mt)')
plt.ylabel('Temperatura (°C)')
plt.title('Correlacion CO₂ vs Temperatura', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)

plt.text(0.05, 0.95, f'r = {corr:.3f}\np = {pval:.4f}', 
         transform=plt.gca().transAxes, bbox=dict(boxstyle="round", facecolor='wheat'),
         verticalalignment='top', fontsize=10)

plt.subplot(2,3,3)
years = df_corr.index.values

slope_co2, intercept_co2, r_co2, p_co2, se_co2 = stats.linregress(years, df_corr['co2'])
trend_co2 = slope_co2 * years + intercept_co2

plt.plot(years, df_corr['co2'], 'bo-', alpha=0.7, label='CO₂ datos')
plt.plot(years, trend_co2, 'b--', linewidth=2, 
         label=f'Tendencia CO₂ ({slope_co2:.3f} Mt/ano)')
plt.xlabel('Ano')
plt.ylabel('CO₂ (Mt)')
plt.title('Tendencia de Emisiones CO₂', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(2,3,4)
slope_temp, intercept_temp, r_temp, p_temp, se_temp = stats.linregress(years, df_corr['temp'])
trend_temp = slope_temp * years + intercept_temp

plt.plot(years, df_corr['temp'], 'ro-', alpha=0.7, label='Temperatura datos')
plt.plot(years, trend_temp, 'r--', linewidth=2, 
         label=f'Tendencia Temp ({slope_temp:.4f}°C/ano)')
plt.xlabel('Ano')
plt.ylabel('Temperatura (°C)')
plt.title('Tendencia de Temperatura', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(2,3,5)
corr_matrix = df_corr.corr()
sns.heatmap(corr_matrix, annot=True, cmap='RdYlBu_r', center=0, 
           square=True, cbar_kws={'label': 'Coeficiente de correlacion'})
plt.title('Matriz de Correlacion', fontsize=14, fontweight='bold')

plt.subplot(2,3,6)
residuos = df_corr['temp'] - y_pred.flatten()
plt.scatter(y_pred, residuos, alpha=0.7, s=60)
plt.axhline(y=0, color='red', linestyle='--', linewidth=2)
plt.xlabel('Valores predichos (°C)')
plt.ylabel('Residuos (°C)')
plt.title('Analisis de Residuos', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\nRESULTADOS ESTADISTICOS DETALLADOS")
print(f"CORRELACION:")
print(f"   • Coeficiente de Pearson: {corr:.4f}")
print(f"   • Valor p: {pval:.2e}")
print(f"   • Significancia: {'Significativa' if pval < 0.05 else 'No significativa'} (α = 0.05)")

print(f"\nTENDENCIAS INDIVIDUALES:")
print(f"   • CO₂: {slope_co2:+.4f} Mt/ano (R² = {r_co2**2:.3f}, p = {p_co2:.2e})")
print(f"   • Temperatura: {slope_temp:+.4f}°C/ano (R² = {r_temp**2:.3f}, p = {p_temp:.2e})")

print(f"\nMODELO DE REGRESION:")
print(f"   • R²: {reg.score(X,y):.4f}")
print(f"   • Pendiente: {reg.coef_[0]:.4f}°C por Mt CO₂")
print(f"   • Intercepto: {reg.intercept_:.2f}°C")
print(f"   • RMSE: {np.sqrt(np.mean(residuos**2)):.4f}°C")

if abs(corr) >= 0.7:
    fuerza = "FUERTE"
    emoji = ""
elif abs(corr) >= 0.5:
    fuerza = "MODERADA"
    emoji = ""
elif abs(corr) >= 0.3:
    fuerza = "DEBIL"
    emoji = ""
else:
    fuerza = "MUY DEBIL"
    emoji = ""

direccion = "POSITIVA" if corr > 0 else "NEGATIVA"

print(f"\n{emoji} INTERPRETACION:")
print(f"   La correlacion es {fuerza} y {direccion}")
if corr > 0:
    print(f"   A mayor emision de CO₂, mayor temperatura")
    print(f"   Por cada 1 Mt de CO₂ adicional, la temperatura aumenta ~{reg.coef_[0]:.3f}°C")
else:
    print(f"   A mayor emision de CO₂, menor temperatura (contraictorio)")

print(f"\nPROYECCIONES (basadas en el modelo):")
co2_scenarios = [1.5, 2.0, 2.5, 3.0]
for co2_val in co2_scenarios:
    temp_pred = reg.predict([[co2_val]])[0]
    print(f"   • {co2_val} Mt CO₂ → {temp_pred:.2f}°C")

print(f"\nLIMITACIONES:")
print(f"   • Correlacion no implica causalidad")
print(f"   • Factores climaticos adicionales no considerados")
print(f"   • Modelo lineal puede ser simplificado")
print(f"   • Muestra relativamente pequena ({len(df_corr)} anos)")
```

### Correlacion: Areas protegidas vs. perdida de biodiversidad
Se analiza si la creacion de areas protegidas se asocia con una reduccion en la perdida de cobertura forestal y/o especies en peligro.

### Modelo predictivo: Regresion lineal para temperatura a partir de emisiones de CO₂
Se entrena un modelo de regresion lineal simple para predecir la temperatura promedio anual en funcion de las emisiones de CO₂.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.preprocessing import PolynomialFeatures
import numpy as np
import joblib

print(f"MODELOS PREDICTIVOS: TEMPERATURA A PARTIR DE EMISIONES CO₂")

if 'df_corr' in locals() and not df_corr.empty:
    print(f"PREPARACION DE DATOS:")
    print(f"   • Periodo: {df_corr.index.min()}-{df_corr.index.max()}")
    print(f"   • Numero de observaciones: {len(df_corr)}")
    print(f"   • Variables: CO₂ (predictora) → Temperatura (objetivo)")
    
    X = df_corr['co2'].values.reshape(-1,1)
    y = df_corr['temp'].values
    
    if len(df_corr) >= 10:
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
        print(f"   • Datos de entrenamiento: {len(X_train)} observaciones")
        print(f"   • Datos de prueba: {len(X_test)} observaciones")
        usar_division = True
    else:
        X_train, X_test, y_train, y_test = X, X, y, y
        print(f"   • Datos insuficientes para division: usando conjunto completo")
        usar_division = False
    
    plt.figure(figsize=(20,15))
    
    print(f"\nMODELO 1: REGRESION LINEAL SIMPLE")
    
    model_linear = LinearRegression()
    model_linear.fit(X_train, y_train)
    
    y_pred_linear_train = model_linear.predict(X_train)
    y_pred_linear_test = model_linear.predict(X_test)
    y_pred_linear_all = model_linear.predict(X)
    
    r2_linear_train = r2_score(y_train, y_pred_linear_train)
    r2_linear_test = r2_score(y_test, y_pred_linear_test)
    rmse_linear = np.sqrt(mean_squared_error(y_test, y_pred_linear_test))
    mae_linear = mean_absolute_error(y_test, y_pred_linear_test)
    
    print(f"   • R² entrenamiento: {r2_linear_train:.4f}")
    print(f"   • R² prueba: {r2_linear_test:.4f}")
    print(f"   • RMSE: {rmse_linear:.4f}°C")
    print(f"   • MAE: {mae_linear:.4f}°C")
    print(f"   • Pendiente: {model_linear.coef_[0]:.4f}°C por Mt CO₂")
    print(f"   • Intercepto: {model_linear.intercept_:.2f}°C")
    
    print(f"\nMODELO 2: REGRESION POLINOMIAL (GRADO 2)")
    
    poly_features = PolynomialFeatures(degree=2)
    X_poly_train = poly_features.fit_transform(X_train)
    X_poly_test = poly_features.transform(X_test)
    X_poly_all = poly_features.transform(X)
    
    model_poly = LinearRegression()
    model_poly.fit(X_poly_train, y_train)
    
    y_pred_poly_train = model_poly.predict(X_poly_train)
    y_pred_poly_test = model_poly.predict(X_poly_test)
    y_pred_poly_all = model_poly.predict(X_poly_all)
    
    r2_poly_train = r2_score(y_train, y_pred_poly_train)
    r2_poly_test = r2_score(y_test, y_pred_poly_test)
    rmse_poly = np.sqrt(mean_squared_error(y_test, y_pred_poly_test))
    mae_poly = mean_absolute_error(y_test, y_pred_poly_test)
    
    print(f"   • R² entrenamiento: {r2_poly_train:.4f}")
    print(f"   • R² prueba: {r2_poly_test:.4f}")
    print(f"   • RMSE: {rmse_poly:.4f}°C")
    print(f"   • MAE: {mae_poly:.4f}°C")
    
    plt.subplot(3,4,1)
    plt.scatter(X, y, s=80, alpha=0.7, c='blue', label='Datos reales', zorder=3)
    
    if usar_division:
        plt.scatter(X_train, y_train, s=60, alpha=0.8, c='green', label='Entrenamiento', zorder=4)
        plt.scatter(X_test, y_test, s=60, alpha=0.8, c='red', marker='s', label='Prueba', zorder=4)
    
    X_smooth = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
    y_linear_smooth = model_linear.predict(X_smooth)
    X_poly_smooth = poly_features.transform(X_smooth)
    y_poly_smooth = model_poly.predict(X_poly_smooth)
    
    plt.plot(X_smooth, y_linear_smooth, 'r-', linewidth=2, label=f'Lineal (R²={r2_linear_test:.3f})')
    plt.plot(X_smooth, y_poly_smooth, 'orange', linewidth=2, linestyle='--', label=f'Polinomial (R²={r2_poly_test:.3f})')
    
    plt.xlabel('Emisiones CO₂ (Mt)')
    plt.ylabel('Temperatura (°C)')
    plt.title('Comparacion de modelos predictivos', fontsize=12, fontweight='bold')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.subplot(3,4,2)
    residuos_linear = y - y_pred_linear_all
    plt.scatter(y_pred_linear_all, residuos_linear, alpha=0.7, s=60, c='red')
    plt.axhline(y=0, color='black', linestyle='--', linewidth=2)
    plt.xlabel('Valores predichos (°C)')
    plt.ylabel('Residuos (°C)')
    plt.title('Residuos - Modelo Lineal', fontsize=12, fontweight='bold')
    plt.grid(True, alpha=0.3)
    
    plt.subplot(3,4,3)
    residuos_poly = y - y_pred_poly_all
    plt.scatter(y_pred_poly_all, residuos_poly, alpha=0.7, s=60, c='orange')
    plt.axhline(y=0, color='black', linestyle='--', linewidth=2)
    plt.xlabel('Valores predichos (°C)')
    plt.ylabel('Residuos (°C)')
    plt.title('Residuos - Modelo Polinomial', fontsize=12, fontweight='bold')
    plt.grid(True, alpha=0.3)
    
    plt.subplot(3,4,4)
    modelos = ['Lineal', 'Polinomial']
    r2_scores = [r2_linear_test, r2_poly_test]
    rmse_scores = [rmse_linear, rmse_poly]
    
    x_pos = np.arange(len(modelos))
    width = 0.35
    
    bars1 = plt.bar(x_pos - width/2, r2_scores, width, label='R²', alpha=0.7, color='skyblue')
    bars2 = plt.bar(x_pos + width/2, [r/10 for r in rmse_scores], width, label='RMSE/10', alpha=0.7, color='lightcoral')
    
    plt.xlabel('Modelo')
    plt.ylabel('Valor de metrica')
    plt.title('Comparacion de rendimiento', fontsize=12, fontweight='bold')
    plt.xticks(x_pos, modelos)
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    for bar, value in zip(bars1, r2_scores):
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
                f'{value:.3f}', ha='center', va='bottom', fontsize=9)
    
    plt.subplot(3,4,5)
    years = df_corr.index.values
    plt.plot(years, y, 'bo-', linewidth=2, markersize=6, label='Temperatura real')
    plt.plot(years, y_pred_linear_all, 'r--', linewidth=2, label='Prediccion lineal')
    plt.plot(years, y_pred_poly_all, 'orange', linestyle=':', linewidth=2, label='Prediccion polinomial')
    plt.xlabel('Ano')
    plt.ylabel('Temperatura (°C)')
    plt.title('Serie temporal: Real vs Predicciones', fontsize=12, fontweight='bold')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.subplot(3,4,6)
    from scipy import stats
    
    def prediction_interval(X_new, X_train, y_train, model, alpha=0.05):
        y_pred = model.predict(X_new)
        
        y_train_pred = model.predict(X_train)
        residuos = y_train - y_train_pred
        mse = np.mean(residuos**2)
        
        df = len(y_train) - 2
        t_val = stats.t.ppf(1 - alpha/2, df)
        
        se = np.sqrt(mse)
        margin = t_val * se
        
        return y_pred - margin, y_pred + margin
    
    X_plot = np.linspace(X.min(), X.max(), 50).reshape(-1, 1)
    y_pred_plot = model_linear.predict(X_plot)
    y_lower, y_upper = prediction_interval(X_plot, X_train, y_train, model_linear)
    
    plt.scatter(X, y, s=60, alpha=0.7, c='blue', label='Datos reales')
    plt.plot(X_plot, y_pred_plot, 'r-', linewidth=2, label='Prediccion')
    plt.fill_between(X_plot.flatten(), y_lower, y_upper, alpha=0.2, color='red', label='IC 95%')
    plt.xlabel('Emisiones CO₂ (Mt)')
    plt.ylabel('Temperatura (°C)')
    plt.title('Intervalos de confianza', fontsize=12, fontweight='bold')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.subplot(3,4,7)
    if len(df_corr) >= 5:
        cv_scores_linear = cross_val_score(model_linear, X, y, cv=min(5, len(df_corr)), scoring='r2')
        cv_scores_poly = cross_val_score(model_poly, X_poly_all, y, cv=min(5, len(df_corr)), scoring='r2')
        
        plt.boxplot([cv_scores_linear, cv_scores_poly], labels=['Lineal', 'Polinomial'])
        plt.ylabel('R² (Validacion cruzada)')
        plt.title('Estabilidad de modelos (CV)', fontsize=12, fontweight='bold')
        plt.grid(True, alpha=0.3)
        
        print(f"\nVALIDACION CRUZADA:")
        print(f"   • Modelo lineal: R² = {cv_scores_linear.mean():.3f} ± {cv_scores_linear.std():.3f}")
        print(f"   • Modelo polinomial: R² = {cv_scores_poly.mean():.3f} ± {cv_scores_poly.std():.3f}")
    else:
        plt.text(0.5, 0.5, 'Datos insuficientes\npara validacion cruzada', 
                ha='center', va='center', transform=plt.gca().transAxes)
        plt.title('Validacion cruzada (N/A)', fontsize=12, fontweight='bold')
    
    plt.subplot(3,4,8)
    
    co2_escenarios = np.array([0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5]).reshape(-1, 1)
    temp_linear_escenarios = model_linear.predict(co2_escenarios)
    
    co2_poly_escenarios = poly_features.transform(co2_escenarios)
    temp_poly_escenarios = model_poly.predict(co2_poly_escenarios)
    
    plt.plot(co2_escenarios, temp_linear_escenarios, 'ro-', linewidth=2, label='Proyeccion lineal')
    plt.plot(co2_escenarios, temp_poly_escenarios, 'o--', color='orange', linewidth=2, label='Proyeccion polinomial')
    
    plt.scatter(X, y, s=40, alpha=0.5, c='blue', label='Datos historicos')
    
    plt.xlabel('Emisiones CO₂ (Mt)')
    plt.ylabel('Temperatura proyectada (°C)')
    plt.title('Proyecciones de temperatura', fontsize=12, fontweight='bold')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    print(f"\nSELECCION DEL MEJOR MODELO:")
    if r2_poly_test > r2_linear_test and rmse_poly < rmse_linear:
        mejor_modelo = "POLINOMIAL"
        mejor_r2 = r2_poly_test
        mejor_rmse = rmse_poly
        modelo_final = model_poly
        features_final = poly_features
        print(f"   Modelo polinomial es superior en ambas metricas")
    elif r2_linear_test > r2_poly_test and rmse_linear < rmse_poly:
        mejor_modelo = "LINEAL"
        mejor_r2 = r2_linear_test
        mejor_rmse = rmse_linear
        modelo_final = model_linear
        features_final = None
        print(f"   Modelo lineal es superior en ambas metricas")
    else:
        diferencia_r2 = abs(r2_poly_test - r2_linear_test)
        if diferencia_r2 < 0.05:
            mejor_modelo = "LINEAL"
            mejor_r2 = r2_linear_test
            mejor_rmse = rmse_linear
            modelo_final = model_linear
            features_final = None
            print(f"   Performance similar, se elige el modelo mas simple (LINEAL)")
        else:
            if r2_poly_test > r2_linear_test:
                mejor_modelo = "POLINOMIAL"
                mejor_r2 = r2_poly_test
                mejor_rmse = rmse_poly
                modelo_final = model_poly
                features_final = poly_features
                print(f"   Modelo polinomial tiene mejor R²")
            else:
                mejor_modelo = "LINEAL"
                mejor_r2 = r2_linear_test
                mejor_rmse = rmse_linear
                modelo_final = model_linear
                features_final = None
                print(f"   Modelo lineal tiene mejor R²")
    
    print(f"\nRESUMEN DEL MODELO SELECCIONADO ({mejor_modelo}):")
    print(f"   • R²: {mejor_r2:.4f}")
    print(f"   • RMSE: {mejor_rmse:.4f}°C")
    print(f"   • MAE: {mae_linear if mejor_modelo == 'LINEAL' else mae_poly:.4f}°C")
    
    model_filename = '../scripts/modelo_temperatura_co2.joblib'
    if features_final is not None:
        joblib.dump({'model': modelo_final, 'poly_features': features_final, 'type': 'polynomial'}, model_filename)
    else:
        joblib.dump({'model': modelo_final, 'poly_features': None, 'type': 'linear'}, model_filename)
    
    print(f"   • Modelo guardado en: {model_filename}")
    
    print(f"\nPREDICCIONES PARA ESCENARIOS ESPECIFICOS:")
    escenarios = [
        (1.0, "Emisiones bajas"),
        (1.5, "Emisiones actuales promedio"),
        (2.0, "Emisiones moderadas"),
        (2.5, "Emisiones altas"),
        (3.0, "Emisiones muy altas")
    ]
    
    for co2_val, descripcion in escenarios:
        if features_final is not None:
            X_pred = features_final.transform([[co2_val]])
            temp_pred = modelo_final.predict(X_pred)[0]
        else:
            temp_pred = modelo_final.predict([[co2_val]])[0]
        
        print(f"   - {co2_val} Mt CO2 ({descripcion}): {temp_pred:.2f}°C")
    
    print(f"\nLIMITACIONES DEL MODELO:")
    print(f"   - Modelo basado solo en datos de Honduras")
    print(f"   - No considera otros gases de efecto invernadero")
    print(f"   - No incluye factores climaticos externos (ENSO, erupciones, etc.)")
    print(f"   - Muestra pequena ({len(df_corr)} anos)")
    print(f"   - Extrapolacion fuera del rango de datos puede ser poco confiable")
    print(f"   - Correlacion no implica causalidad directa")
    
else:
    print('ERROR: No hay datos suficientes para entrenar el modelo.')
    print('   Verifique que el dataframe df_corr este disponible y contenga datos.')
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import numpy as np
import pandas as pd

print(f"MODELO PREDICTIVO MULTIFACTORIAL: DEFORESTACION")

try:
    years_deforest = [int(y) for y in df_loss_hnd.index]
    
    datos_modelo = pd.DataFrame({
        'año': years_deforest,
        'deforestacion': total_loss,
    })
    
    datos_modelo['año_normalizado'] = (datos_modelo['año'] - datos_modelo['año'].min()) / (datos_modelo['año'].max() - datos_modelo['año'].min())
    datos_modelo['tendencia_temporal'] = np.arange(len(datos_modelo))
    
    if 'df_corr' in locals() and not df_corr.empty:
        co2_reindexed = df_corr['co2'].reindex(years_deforest, method='nearest').fillna(method='ffill').fillna(method='bfill')
        datos_modelo['co2'] = co2_reindexed.values
        print(f"Variable CO2 agregada ({len(co2_reindexed.dropna())} valores)")
    else:
        datos_modelo['co2'] = np.linspace(1.0, 2.0, len(datos_modelo))
        print(f"Variable CO2 sintetica creada")
    
    if 'df_temp_anual' in locals() and not df_temp_anual.empty:
        temp_reindexed = df_temp_anual.set_index('Year')['Estimated_AverageTemperature_C'].reindex(years_deforest, method='nearest').fillna(method='ffill').fillna(method='bfill')
        datos_modelo['temperatura'] = temp_reindexed.values
        print(f"Variable temperatura agregada ({len(temp_reindexed.dropna())} valores)")
    else:
        datos_modelo['temperatura'] = 25 + np.random.normal(0, 1, len(datos_modelo))
        print(f"Variable temperatura sintetica creada")
    
    if 'areas_acum' in locals():
        datos_modelo['areas_protegidas'] = areas_acum.values
        print(f"Variable areas protegidas agregada")
    else:
        datos_modelo['areas_protegidas'] = np.cumsum(np.random.poisson(2, len(datos_modelo)))
        print(f"Variable areas protegidas sintetica creada")
    
    datos_modelo['co2_lag1'] = datos_modelo['co2'].shift(1).fillna(datos_modelo['co2'].iloc[0])
    datos_modelo['temp_lag1'] = datos_modelo['temperatura'].shift(1).fillna(datos_modelo['temperatura'].iloc[0])
    datos_modelo['deforest_lag1'] = datos_modelo['deforestacion'].shift(1).fillna(datos_modelo['deforestacion'].iloc[0])
    
    datos_modelo['co2_x_temp'] = datos_modelo['co2'] * datos_modelo['temperatura']
    datos_modelo['areas_x_año'] = datos_modelo['areas_protegidas'] * datos_modelo['año_normalizado']
    
    print(f"\nDATASET PREPARADO:")
    print(f"   • Numero de observaciones: {len(datos_modelo)}")
    print(f"   • Periodo: {min(years_deforest)}-{max(years_deforest)}")
    print(f"   • Variables predictoras: {len(datos_modelo.columns) - 1}")
    
    print(f"\nESTADISTICAS DESCRIPTIVAS:")
    display(datos_modelo.describe().round(2))
    
    feature_columns = [col for col in datos_modelo.columns if col not in ['año', 'deforestacion']]
    X = datos_modelo[feature_columns]
    y = datos_modelo['deforestacion']
    
    print(f"\nVARIABLES PREDICTORAS SELECCIONADAS:")
    for i, col in enumerate(feature_columns):
        print(f"   {i+1}. {col}")
    
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    X_scaled_df = pd.DataFrame(X_scaled, columns=feature_columns)
    
    if len(datos_modelo) >= 8:
        X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
        usar_division = True
        print(f"\nDIVISION DE DATOS:")
        print(f"   - Entrenamiento: {len(X_train)} observaciones")
        print(f"   - Prueba: {len(X_test)} observaciones")
    else:
        X_train, X_test, y_train, y_test = X_scaled, X_scaled, y, y
        usar_division = False
        print(f"\nDATOS INSUFICIENTES para division: usando conjunto completo")
    
    print(f"\nMODELO 1: REGRESION LINEAL MULTIPLE")
    
    from sklearn.linear_model import LinearRegression
    model_linear_multi = LinearRegression()
    model_linear_multi.fit(X_train, y_train)
    
    y_pred_linear_train = model_linear_multi.predict(X_train)
    y_pred_linear_test = model_linear_multi.predict(X_test)
    y_pred_linear_all = model_linear_multi.predict(X_scaled)
    
    r2_linear_multi_train = r2_score(y_train, y_pred_linear_train)
    r2_linear_multi_test = r2_score(y_test, y_pred_linear_test)
    rmse_linear_multi = np.sqrt(mean_squared_error(y_test, y_pred_linear_test))
    mae_linear_multi = mean_absolute_error(y_test, y_pred_linear_test)
    
    print(f"   - R² entrenamiento: {r2_linear_multi_train:.4f}")
    print(f"   - R² prueba: {r2_linear_multi_test:.4f}")
    print(f"   - RMSE: {rmse_linear_multi:.2f} ha")
    print(f"   - MAE: {mae_linear_multi:.2f} ha")
    
    print(f"\nMODELO 2: RANDOM FOREST")
    
    model_rf = RandomForestRegressor(n_estimators=100, random_state=42, max_depth=5)
    model_rf.fit(X_train, y_train)
    
    y_pred_rf_train = model_rf.predict(X_train)
    y_pred_rf_test = model_rf.predict(X_test)
    y_pred_rf_all = model_rf.predict(X_scaled)
    
    r2_rf_train = r2_score(y_train, y_pred_rf_train)
    r2_rf_test = r2_score(y_test, y_pred_rf_test)
    rmse_rf = np.sqrt(mean_squared_error(y_test, y_pred_rf_test))
    mae_rf = mean_absolute_error(y_test, y_pred_rf_test)
    
    print(f"   - R² entrenamiento: {r2_rf_train:.4f}")
    print(f"   - R² prueba: {r2_rf_test:.4f}")
    print(f"   - RMSE: {rmse_rf:.2f} ha")
    print(f"   - MAE: {mae_rf:.2f} ha")
    
    feature_importance = model_rf.feature_importances_
    importance_df = pd.DataFrame({
        'Variable': feature_columns,
        'Importancia': feature_importance
    }).sort_values('Importancia', ascending=False)
    
    print(f"\nIMPORTANCIA DE VARIABLES (Random Forest):")
    for idx, row in importance_df.iterrows():
        print(f"   - {row['Variable']}: {row['Importancia']:.3f}")
    
    plt.figure(figsize=(20, 15))
    
    plt.subplot(3, 4, 1)
    plt.scatter(y, y_pred_linear_all, alpha=0.7, s=80, label='Lineal multiple', color='blue')
    plt.scatter(y, y_pred_rf_all, alpha=0.7, s=80, label='Random Forest', color='green')
    plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--', linewidth=2, label='Prediccion perfecta')
    plt.xlabel('Deforestacion real (ha)')
    plt.ylabel('Deforestacion predicha (ha)')
    plt.title('Predicciones vs Realidad', fontsize=12, fontweight='bold')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.subplot(3, 4, 2)
    plt.plot(years_deforest, y, 'ko-', linewidth=2, markersize=6, label='Real')
    plt.plot(years_deforest, y_pred_linear_all, 'b--', linewidth=2, label='Lineal multiple')
    plt.plot(years_deforest, y_pred_rf_all, 'g:', linewidth=2, label='Random Forest')
    plt.xlabel('Ano')
    plt.ylabel('Deforestacion (ha)')
    plt.title('Serie temporal: Real vs Predicciones', fontsize=12, fontweight='bold')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.subplot(3, 4, 3)
    plt.barh(importance_df['Variable'], importance_df['Importancia'], alpha=0.7, color='skyblue')
    plt.xlabel('Importancia')
    plt.title('Importancia de variables (RF)', fontsize=12, fontweight='bold')
    plt.grid(True, alpha=0.3)
    
    plt.subplot(3, 4, 4)
    residuos_linear_multi = y - y_pred_linear_all
    plt.scatter(y_pred_linear_all, residuos_linear_multi, alpha=0.7, s=60, color='blue')
    plt.axhline(y=0, color='red', linestyle='--', linewidth=2)
    plt.xlabel('Predicciones (ha)')
    plt.ylabel('Residuos (ha)')
    plt.title('Residuos - Regresion lineal', fontsize=12, fontweight='bold')
    plt.grid(True, alpha=0.3)
    
    plt.subplot(3, 4, 5)
    residuos_rf = y - y_pred_rf_all
    plt.scatter(y_pred_rf_all, residuos_rf, alpha=0.7, s=60, color='green')
    plt.axhline(y=0, color='red', linestyle='--', linewidth=2)
    plt.xlabel('Predicciones (ha)')
    plt.ylabel('Residuos (ha)')
    plt.title('Residuos - Random Forest', fontsize=12, fontweight='bold')
    plt.grid(True, alpha=0.3)
    
    plt.subplot(3, 4, 6)
    modelos = ['Lineal\\nMultiple', 'Random\\nForest']
    r2_scores = [r2_linear_multi_test, r2_rf_test]
    rmse_scores = [rmse_linear_multi, rmse_rf]
    
    x_pos = np.arange(len(modelos))
    width = 0.35
    
    bars1 = plt.bar(x_pos - width/2, r2_scores, width, label='R²', alpha=0.7, color='lightblue')
    bars2 = plt.bar(x_pos + width/2, [r/1000 for r in rmse_scores], width, 
                   label='RMSE/1000', alpha=0.7, color='lightcoral')
    
    plt.xlabel('Modelo')
    plt.ylabel('Valor de metrica')
    plt.title('Comparacion de rendimiento', fontsize=12, fontweight='bold')
    plt.xticks(x_pos, modelos)
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    for bar, value in zip(bars1, r2_scores):
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
                f'{value:.3f}', ha='center', va='bottom', fontsize=9)
    
    plt.subplot(3, 4, 7)
    corr_predictoras = X_scaled_df.corr()
    sns.heatmap(corr_predictoras, annot=True, cmap='RdYlBu_r', center=0, 
               square=True, cbar_kws={'label': 'Correlacion'}, fmt='.2f')
    plt.title('Correlaciones entre predictoras', fontsize=12, fontweight='bold')
    
    plt.subplot(3, 4, 8)
    plt.hist(residuos_linear_multi, bins=6, alpha=0.7, color='blue', label='Lineal', density=True)
    plt.hist(residuos_rf, bins=6, alpha=0.7, color='green', label='Random Forest', density=True)
    plt.xlabel('Residuos (ha)')
    plt.ylabel('Densidad')
    plt.title('Distribucion de residuos', fontsize=12, fontweight='bold')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.subplot(3, 4, 9)
    if len(datos_modelo) >= 5:
        cv_scores_linear = cross_val_score(model_linear_multi, X_scaled, y, cv=min(5, len(datos_modelo)), scoring='r2')
        cv_scores_rf = cross_val_score(model_rf, X_scaled, y, cv=min(5, len(datos_modelo)), scoring='r2')
        
        plt.boxplot([cv_scores_linear, cv_scores_rf], labels=['Lineal', 'Random\\nForest'])
        plt.ylabel('R² (Validacion cruzada)')
        plt.title('Estabilidad de modelos (CV)', fontsize=12, fontweight='bold')
        plt.grid(True, alpha=0.3)
        
        print(f"\nVALIDACION CRUZADA:")
        print(f"   - Regresion lineal: R² = {cv_scores_linear.mean():.3f} ± {cv_scores_linear.std():.3f}")
        print(f"   - Random Forest: R² = {cv_scores_rf.mean():.3f} ± {cv_scores_rf.std():.3f}")
    
    plt.subplot(3, 4, 10)
    variable_importante = importance_df.iloc[0]['Variable']
    plt.scatter(X_scaled_df[variable_importante], y, alpha=0.7, s=60)
    
    from scipy.stats import linregress
    slope, intercept, r_val, p_val, se = linregress(X_scaled_df[variable_importante], y)
    x_trend = np.linspace(X_scaled_df[variable_importante].min(), X_scaled_df[variable_importante].max(), 100)
    y_trend = slope * x_trend + intercept
    plt.plot(x_trend, y_trend, 'r--', linewidth=2, label=f'r = {r_val:.3f}')
    
    plt.xlabel(f'{variable_importante} (estandarizada)')
    plt.ylabel('Deforestacion (ha)')
    plt.title(f'Variable mas importante: {variable_importante}', fontsize=12, fontweight='bold')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.subplot(3, 4, 11)
    
    años_futuros = np.arange(max(years_deforest) + 1, max(years_deforest) + 6)
    
    escenarios_futuros = []
    for año in años_futuros:
        año_norm = (año - min(years_deforest)) / (max(years_deforest) - min(years_deforest))
        tendencia = len(datos_modelo) + (año - max(years_deforest))
        
        co2_fut = datos_modelo['co2'].iloc[-1] + 0.1 * (año - max(years_deforest))
        temp_fut = datos_modelo['temperatura'].iloc[-1] + 0.05 * (año - max(years_deforest))
        areas_fut = datos_modelo['areas_protegidas'].iloc[-1] + 2 * (año - max(years_deforest))
        
        escenario = {
            'año_normalizado': año_norm,
            'tendencia_temporal': tendencia,
            'co2': co2_fut,
            'temperatura': temp_fut,
            'areas_protegidas': areas_fut,
            'co2_lag1': datos_modelo['co2'].iloc[-1],
            'temp_lag1': datos_modelo['temperatura'].iloc[-1],
            'deforest_lag1': datos_modelo['deforestacion'].iloc[-1],
            'co2_x_temp': co2_fut * temp_fut,
            'areas_x_año': areas_fut * año_norm
        }
        escenarios_futuros.append(escenario)
    
    escenarios_df = pd.DataFrame(escenarios_futuros)
    escenarios_scaled = scaler.transform(escenarios_df)
    
    pred_linear_futuro = model_linear_multi.predict(escenarios_scaled)
    pred_rf_futuro = model_rf.predict(escenarios_scaled)
    
    años_completos = list(years_deforest) + list(años_futuros)
    deforest_completa = list(y) + [np.nan] * len(años_futuros)
    pred_linear_completa = list(y_pred_linear_all) + list(pred_linear_futuro)
    pred_rf_completa = list(y_pred_rf_all) + list(pred_rf_futuro)
    
    plt.plot(years_deforest, y, 'ko-', linewidth=2, markersize=6, label='Historico')
    plt.plot(años_completos, pred_linear_completa, 'b--', linewidth=2, label='Proyeccion lineal')
    plt.plot(años_completos, pred_rf_completa, 'g:', linewidth=2, label='Proyeccion RF')
    plt.axvline(x=max(years_deforest), color='red', linestyle=':', alpha=0.7, label='Inicio proyeccion')
    plt.xlabel('Ano')
    plt.ylabel('Deforestacion (ha)')
    plt.title('Proyecciones futuras de deforestacion', fontsize=12, fontweight='bold')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.subplot(3, 4, 12)
    
    n_bootstrap = 100
    predictions_bootstrap = []
    
    for _ in range(n_bootstrap):
        indices = np.random.choice(len(X_train), size=len(X_train), replace=True)
        X_boot = X_train[indices]
        y_boot = y_train.iloc[indices] if hasattr(y_train, 'iloc') else y_train[indices]
        
        model_boot = RandomForestRegressor(n_estimators=50, random_state=None, max_depth=5)
        model_boot.fit(X_boot, y_boot)
        
        pred_boot = model_boot.predict(X_scaled)
        predictions_bootstrap.append(pred_boot)
    
    predictions_bootstrap = np.array(predictions_bootstrap)
    pred_mean = np.mean(predictions_bootstrap, axis=0)
    pred_std = np.std(predictions_bootstrap, axis=0)
    pred_lower = np.percentile(predictions_bootstrap, 2.5, axis=0)
    pred_upper = np.percentile(predictions_bootstrap, 97.5, axis=0)
    
    plt.fill_between(years_deforest, pred_lower, pred_upper, alpha=0.3, color='green', label='IC 95%')
    plt.plot(years_deforest, pred_mean, 'g-', linewidth=2, label='Prediccion media')
    plt.scatter(years_deforest, y, s=60, color='red', label='Datos reales', zorder=5)
    plt.xlabel('Ano')
    plt.ylabel('Deforestacion (ha)')
    plt.title('Incertidumbre del modelo (Bootstrap)', fontsize=12, fontweight='bold')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    print(f"\nSELECCION DEL MEJOR MODELO:")
    
    if r2_rf_test > r2_linear_multi_test and rmse_rf < rmse_linear_multi:
        mejor_modelo_deforest = "RANDOM FOREST"
        mejor_r2_deforest = r2_rf_test
        mejor_rmse_deforest = rmse_rf
        modelo_final_deforest = model_rf
        print(f"   Random Forest es superior en ambas metricas")
    elif r2_linear_multi_test > r2_rf_test and rmse_linear_multi < rmse_rf:
        mejor_modelo_deforest = "REGRESION LINEAL MULTIPLE"
        mejor_r2_deforest = r2_linear_multi_test
        mejor_rmse_deforest = rmse_linear_multi
        modelo_final_deforest = model_linear_multi
        print(f"   Regresion lineal multiple es superior en ambas metricas")
    else:
        if r2_rf_test > r2_linear_multi_test:
            mejor_modelo_deforest = "RANDOM FOREST"
            mejor_r2_deforest = r2_rf_test
            mejor_rmse_deforest = rmse_rf
            modelo_final_deforest = model_rf
            print(f"   Random Forest tiene mejor R²")
        else:
            mejor_modelo_deforest = "REGRESION LINEAL MULTIPLE"
            mejor_r2_deforest = r2_linear_multi_test
            mejor_rmse_deforest = rmse_linear_multi
            modelo_final_deforest = model_linear_multi
            print(f"   Regresion lineal tiene mejor R²")
    
    print(f"\nRESUMEN DEL MODELO SELECCIONADO ({mejor_modelo_deforest}):")
    print(f"   - R²: {mejor_r2_deforest:.4f}")
    print(f"   - RMSE: {mejor_rmse_deforest:.2f} ha")
    print(f"   - MAE: {mae_rf if mejor_modelo_deforest == 'RANDOM FOREST' else mae_linear_multi:.2f} ha")
    
    model_deforest_filename = '../scripts/modelo_deforestacion_multifactorial.joblib'
    joblib.dump({
        'model': modelo_final_deforest,
        'scaler': scaler,
        'feature_columns': feature_columns,
        'type': mejor_modelo_deforest.lower().replace(' ', '_'),
        'importance': importance_df if mejor_modelo_deforest == 'RANDOM FOREST' else None
    }, model_deforest_filename)
    
    print(f"   - Modelo guardado en: {model_deforest_filename}")
    
    print(f"\nINTERPRETABILIDAD DEL MODELO:")
    
    if mejor_modelo_deforest == "RANDOM FOREST":
        print(f"   TOP 3 VARIABLES MAS IMPORTANTES:")
        for i in range(min(3, len(importance_df))):
            var = importance_df.iloc[i]['Variable']
            imp = importance_df.iloc[i]['Importancia']
            print(f"      {i+1}. {var}: {imp:.3f}")
        
        var_top = importance_df.iloc[0]['Variable']
        corr_top = np.corrcoef(X_scaled_df[var_top], y)[0,1]
        print(f"\n   ANALISIS DE VARIABLE MAS IMPORTANTE ({var_top}):")
        print(f"      • Correlacion con deforestacion: {corr_top:.3f}")
        print(f"      • Importancia relativa: {importance_df.iloc[0]['Importancia']:.3f}")
        
        if 'areas_protegidas' in var_top:
            print(f"      • Interpretacion: Las areas protegidas {'reducen' if corr_top < 0 else 'se asocian con mayor'} deforestacion")
        elif 'co2' in var_top:
            print(f"      • Interpretacion: Mayor CO₂ {'reduce' if corr_top < 0 else 'aumenta'} la deforestacion")
        elif 'temp' in var_top:
            print(f"      • Interpretacion: Mayor temperatura {'reduce' if corr_top < 0 else 'aumenta'} la deforestacion")
    
    else:
        coeficientes = model_linear_multi.coef_
        coef_df = pd.DataFrame({
            'Variable': feature_columns,
            'Coeficiente': coeficientes,
            'Coef_Abs': np.abs(coeficientes)
        }).sort_values('Coef_Abs', ascending=False)
        
        print(f"   COEFICIENTES MAS IMPORTANTES:")
        for i in range(min(3, len(coef_df))):
            var = coef_df.iloc[i]['Variable']
            coef = coef_df.iloc[i]['Coeficiente']
            print(f"      {i+1}. {var}: {coef:+.3f}")
            
            if coef > 0:
                print(f"         -> Mayor {var} aumenta la deforestacion")
            else:
                print(f"         -> Mayor {var} reduce la deforestacion")
    
    print(f"\nPREDICCIONES PARA ESCENARIOS ESPECIFICOS:")
    
    print(f"\n   ESCENARIO 1: Mejora ambiental")
    print(f"      - Reduccion 20% CO2, +10% areas protegidas, temperatura estable")
    
    print(f"\n   ESCENARIO 2: Deterioro ambiental") 
    print(f"      - Aumento 30% CO2, areas protegidas estables, +1°C temperatura")
    
    print(f"\n   ESCENARIO 3: Politicas de conservacion")
    print(f"      - CO2 estable, +50% areas protegidas, temperatura estable")
    
    base_data = datos_modelo.iloc[-1].copy()
    
    escenarios_nombres = ['Mejora ambiental', 'Deterioro ambiental', 'Politicas conservacion']
    modificadores_co2 = [0.8, 1.3, 1.0]
    modificadores_areas = [1.1, 1.0, 1.5]
    modificadores_temp = [1.0, 1.04, 1.0]
    
    for i, (nombre, mod_co2, mod_areas, mod_temp) in enumerate(zip(escenarios_nombres, modificadores_co2, modificadores_areas, modificadores_temp)):
        escenario = base_data.copy()
        escenario['co2'] *= mod_co2
        escenario['areas_protegidas'] *= mod_areas
        escenario['temperatura'] *= mod_temp
        escenario['co2_x_temp'] = escenario['co2'] * escenario['temperatura']
        escenario['areas_x_año'] = escenario['areas_protegidas'] * escenario['año_normalizado']
        
        escenario_vector = escenario[feature_columns].values.reshape(1, -1)
        escenario_scaled = scaler.transform(escenario_vector)
        
        pred_escenario = modelo_final_deforest.predict(escenario_scaled)[0]
        
        print(f"\n   {i+1}. {nombre}:")
        print(f"      - Deforestacion predicha: {pred_escenario:.0f} ha/ano")
        print(f"      - Cambio vs. historico promedio: {((pred_escenario - np.mean(y)) / np.mean(y)) * 100:+.1f}%")
    
    print(f"\nLIMITACIONES DEL MODELO:")
    print(f"   - Muestra pequena ({len(datos_modelo)} observaciones)")
    print(f"   - Posible sobreajuste debido a pocas observaciones")
    print(f"   - Variables sinteticas usadas donde faltan datos reales")
    print(f"   - No considera factores socioeconomicos")
    print(f"   - No incluye eventos extremos (sequias, huracanes)")
    print(f"   - Relaciones causales no establecidas")
    
    print(f"\nRECOMENDACIONES:")
    print(f"   1. Recopilar mas datos historicos para mejorar robustez")
    print(f"   2. Incluir variables socioeconomicas (PIB, poblacion, precios)")
    print(f"   3. Considerar efectos de rezago temporal")
    print(f"   4. Validar predicciones con datos independientes")
    print(f"   5. Actualizar modelo regularmente con nuevos datos")
    print(f"   6. Complementar con estudios de campo y teledeteccion")
    
    print(f"\nMODELO COMPLETADO EXITOSAMENTE")
    
except Exception as e:
    print(f"ERROR en la creacion del modelo multifactorial:")
    print(f"   {str(e)}")
    print(f"   Esto puede deberse a datos insuficientes o inconsistencias en los datasets")
```
